```python
import hail as hl
from identify_ancestry import AncestryPredictor

%load_ext autoreload
%autoreload 2
```

```python
data_path = 'data/hgdp'
mt = hl.read_matrix_table(f'{data_path}/HGDP.mt')
```

```python
populations_table = hl.import_table(f'{data_path}/HGDP_annotations.txt', key='s')
populations_table.show()
```

```python
mt = mt.annotate_cols(**populations_table[mt.s])
```

```python
hl.summarize_variants(mt)
```

```python
import random
import pandas as pd

rsid_list = mt.aggregate_rows(hl.agg.collect_as_set(mt.rsid))
selected_snps = random.choices(list(rsid_list), k=10)
selected_snps
```

```python
predictor = AncestryPredictor(mt)
```

```python
mt_probs = predictor.fit(selected_snps)
```

```python
predictor.predict_ancestry(mt_probs).cols().show()
```
